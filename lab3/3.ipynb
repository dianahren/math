# Множення матриць

```python
A = np.array([ [1, 2, 5], [2, 2, 2], [5, 5, 5] ])
B = np.array([ [5, 2, 1], [1, 2, 4], [-1, -2, -4] ])
print(f"Маємо два масиви:\n {A} \n\n {B}")

R = A * B
print(f"\nТут виконується поелементне множення:\n {R}")

R = np.dot(A, B)
print(f"\nТут, масиви перемножуються як матриці:\n {R}")

MA = np.mat(A)
MB = np.mat(B)

R = MA * MB
print(f"\nТут масиви перетворені на матриці і далі перемножуються:\n {R}")
```
